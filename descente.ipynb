# génération valeurs


```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
n = 100
x_train= np.random.rand(n) * 10  
true_m = 2.5
true_b = 1.0
noise = np.random.randn(n)  
y_train = true_m * x_train + true_b + noise

# Plot the data
plt.scatter(x_train, y_train)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Synthetic Data: y = 2.5 x + 1 + noise")
plt.grid()
plt.show()


```

# Fonction de perte

```python
def loss_function(m, b, x, y):    # en gros  on prend x et y comme variable comme ca la fonction marche pour toute donnée qu'on souhaite; par exemple ici c'est _train mais pourrais prendre x_trainversion2
        n = len(x)
        loss = 0
        for i in range(n):
            y_pred = m * x[i]+b
            loss = loss +  (y[i]-y_pred)**2
        return loss/n
```

# calcule du gradient

```python
def compute_gradients(m, b, x, y):
    
    z = len(x)
    dm = 0
    db = 0
    gradient_b = 0
    gradient_m = 0
    for i in range(z):
        ypred = m * x[i] + b
        errorg = ypred - y[i]
        dm = dm + errorg * x[i]
        db = db + errorg
        
    gradient_m = dm / z
    gradient_b = db / z

    return gradient_m, gradient_b
    

```

# entrainement du modèle

```python
def trainmodel(x , y ,m, b, alpha , epochs):
    for epoch in range(epochs):
        gradient_m, gradient_b = compute_gradients(m, b, x, y)
        m = m - alpha * gradient_m
        b = b - alpha * gradient_b
        error = loss_function(m, b, x, y)
    return m, b, error




m , b, error = trainmodel(x_train, y_train, 0.0, 0.0, 0.001, 10000)

print(m, b, error)


```
